## Pembentukan Variabel *Fuzzy*

```python
# Definisikan nilai Min, Max, dan Kuartil Data
min = {
    'permintaan'            : 8,
    'persediaan'            : 618,
    'harga_bahan_baku'      : 56650,
    'tenaga_kerja'          : 4,
    'jumlah_produksi_aktual': 62
}

max = {
    'permintaan'            : 436,
    'persediaan'            : 1046,
    'harga_bahan_baku'      : 64950,
    'tenaga_kerja'          : 15,
    'jumlah_produksi_aktual': 490
}

kuartil = {
    'permintaan'            : [32, 76, 140],
    'persediaan'            : [914, 978, 1022],
    'harga_bahan_baku'      : [57120, 57980, 59200],
    'tenaga_kerja'          : [5, 8, 12],
    'jumlah_produksi_aktual': [86, 135, 200]
}

# Definisikan Variabel Fuzzy
x = {
    'permintaan':np.arange(min['permintaan'],
                           max['permintaan']+1, 1,
                           dtype=float),
    'persediaan':np.arange(min['persediaan'],
                           max['persediaan']+1,1,
                           dtype=float),
    'harga_bahan_baku':np.arange(min['harga_bahan_baku'],
                                 max['harga_bahan_baku']+1, 1,
                                 dtype=float),
    'tenaga_kerja':np.arange(min['tenaga_kerja'],
                             max['tenaga_kerja']+1, 1,
                             dtype=float),
    'jumlah_produksi_aktual':np.arange(min['jumlah_produksi_aktual'],
                                       max['jumlah_produksi_aktual']+0.01, 0.01,
                                       dtype=float),
}
```

## Pembentukan Himpunan *Fuzzy*

```python
# Definisikan Himpunan Fuzzy dan fungsi keanggotaan (membership functions)
# Variabel Input
permintaan = {
    'RENDAH':fuzz.trimf(x['permintaan'], [min['permintaan'],
                                          min['permintaan'],
                                          kuartil['permintaan'][1]]),
    'SEDANG':fuzz.trimf(x['permintaan'], [kuartil['permintaan'][0],
                                          kuartil['permintaan'][1],
                                          kuartil['permintaan'][2]]),
    'TINGGI':fuzz.trimf(x['permintaan'], [kuartil['permintaan'][1],
                                          max['permintaan'],
                                          max['permintaan']])
}
persediaan = {
    'SEDIKIT':fuzz.trimf(x['persediaan'], [min['persediaan'],
                                           min['persediaan'],
                                           kuartil['persediaan'][1]]),
    'CUKUP':fuzz.trimf(x['persediaan'], [kuartil['persediaan'][0],
                                         kuartil['persediaan'][ 1],
                                         kuartil['persediaan'][2]]),
    'BANYAK':fuzz.trimf(x['persediaan'], [kuartil['persediaan'][1],
                                          max['persediaan'],
                                          max['persediaan']])
}
harga_bahan_baku = {
    'MURAH':fuzz.trimf(x['harga_bahan_baku'], [min['harga_bahan_baku'],
                                               min['harga_bahan_baku'],
                                               kuartil['harga_bahan_baku'][1]]),
    'NORMAL':fuzz.trimf(x['harga_bahan_baku'], [kuartil['harga_bahan_baku'][0],
                                                kuartil['harga_bahan_baku'][1],
                                                kuartil['harga_bahan_baku'][2]]),
    'MAHAL':fuzz.trimf(x['harga_bahan_baku'], [kuartil['harga_bahan_baku'][1],
                                               max['harga_bahan_baku'],
                                               max['harga_bahan_baku']])
}
tenaga_kerja = {
    'SEDIKIT':fuzz.trimf(x['tenaga_kerja'], [min['tenaga_kerja'],
                                             min['tenaga_kerja'],
                                             kuartil['tenaga_kerja'][1]]),
    'CUKUP':fuzz.trimf(x['tenaga_kerja'], [kuartil['tenaga_kerja'][0],
                                           kuartil['tenaga_kerja'][ 1],
                                           kuartil['tenaga_kerja'][2]]),
    'BANYAK':fuzz.trimf(x['tenaga_kerja'], [kuartil['tenaga_kerja'][1],
                                            max['tenaga_kerja'],
                                            max['tenaga_kerja']])
}

# Variabel Output
jumlah_produksi_aktual = {
    'SEDIKIT':fuzz.trimf(x['jumlah_produksi_aktual'], [min['jumlah_produksi_aktual'],
                                                       min['jumlah_produksi_aktual'],
                                                       kuartil['jumlah_produksi_aktual'][1]]),
    'CUKUP':fuzz.trimf(x['jumlah_produksi_aktual'], [kuartil['jumlah_produksi_aktual'][0],
                                                     kuartil['jumlah_produksi_aktual'][1],
                                                     kuartil['jumlah_produksi_aktual'][2]]),
    'BANYAK':fuzz.trimf(x['jumlah_produksi_aktual'], [kuartil['jumlah_produksi_aktual'][1],
                                                      max['jumlah_produksi_aktual'],
                                                      max['jumlah_produksi_aktual']])
}


fungsi_keanggotaan = {
    'permintaan':permintaan,
    'persediaan':persediaan,
    'harga_bahan_baku':harga_bahan_baku,
    'tenaga_kerja':tenaga_kerja,
    'jumlah_produksi_aktual':jumlah_produksi_aktual
}
```

```python
# Grafik Fungsi Keanggotaan
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 4 * 2))
axes = axes.flatten()

# Plot setiap fungsi keanggotaan
for i, (var_name, memberships) in enumerate(fungsi_keanggotaan.items()):
    ax = axes[i]
    ax.set_title(f'Fungsi Keanggotaan {var_name.replace("_"," ").capitalize()}')
    ax.set_xlabel(var_name.replace("_", " ").capitalize())
    ax.set_ylabel('Derajat Keanggotaan μ(x)')
    ax.set_ylim([0, 1.05])
    for level, mf_values in memberships.items():
        ax.plot(x[var_name], mf_values, label=level)

    ax.vlines(min[var_name], 0, 1, color='black',
              linestyle='--', linewidth=1, alpha=0.5)
    ax.vlines(kuartil[var_name][1], 0, 1, color='black',
              linestyle='--', linewidth=1, alpha=0.5)
    ax.vlines(max[var_name], 0, 1, color='black',
              linestyle='--', linewidth=1, alpha=0.5)
    ax.axhline(y=1, color='black', linestyle='--', linewidth=1, alpha=0.5)


    # Tentukan posisi tick sumbu X
    xticks = [min[var_name]] + kuartil[var_name] + [max[var_name]]

    # Terapkan ke axis
    ax.set_xticks(xticks)
    ax.set_xticklabels([min[var_name],
                        kuartil[var_name][0],
                        kuartil[var_name][1],
                        kuartil[var_name][2],
                        max[var_name]])

    ax.legend()
    ax.grid(False)

plt.tight_layout()
plt.show()
```

## Fuzzifikasi

```python
def fuzzifikasi(input):
    hasil = {}
    for var, nilai in input.items():
        hasil[var] = {}
        for label, mf in fungsi_keanggotaan[var].items():
            derajat = fuzz.interp_membership(x[var], mf, nilai)
            hasil[var][label] = round(derajat,4)
    return hasil
```

## Aplikasi Fungsi Implikasi *Fuzzy*


```python
label_skor1 = {'RENDAH': 0, 'SEDANG': 2, 'TINGGI': 3}
label_skor2 = {'SEDIKIT': 1, 'CUKUP': 0.5, 'BANYAK': 0, 'MURAH':1, 'NORMAL':0.5, 'MAHAL':0}
label_skor_tenagakerja = {'SEDIKIT': 0, 'CUKUP': 1, 'BANYAK': 2}

label_1 = ['RENDAH', 'SEDANG', 'TINGGI']
label_2 = ['SEDIKIT', 'CUKUP', 'BANYAK']
label_3 = ['MURAH', 'NORMAL', 'MAHAL']

i = 1
implikasi = [0,0,0]
MIU = []

print(f"Aplikasi Fungsi Implikasi Fuzzy:")
for pmt in label_1:
    for psd in label_2:
        for bbk in label_3:
            for tkg in label_2:
                # Hitung total skor dari kombinasi label
                # Konsekuen
                if pmt == 'RENDAH':
                    total_score = (
                        label_skor2[psd] + label_skor_tenagakerja[tkg] +
                        label_skor2[bbk]
                    )
                    if total_score <= 3:
                      output = 'SEDIKIT'
                    else:
                      output = 'CUKUP'

                elif pmt == 'SEDANG':
                    total_score = (
                      label_skor1[pmt] + label_skor2[psd] +
                      label_skor2[bbk] + label_skor_tenagakerja[tkg]
                    )
                    if total_score <= 3:
                        output = 'SEDIKIT'
                    elif total_score <= 5:
                        output = 'CUKUP'
                    else:
                        output = 'BANYAK'
                else:
                    total_score = (
                      label_skor1[pmt] + label_skor2[psd] +
                      label_skor2[bbk] + label_skor_tenagakerja[tkg]
                    )
                    if total_score <= 2:
                        output = 'SEDIKIT'
                    elif total_score <= 4:
                        output = 'CUKUP'
                    else:
                        output = 'BANYAK'

                # Aturan
                aturan = [pmt, psd, bbk, tkg, output]
                hasil = hasil_fuzzifikasi

                # Aplikasi fungsi implikasi tiap aturan
                miu = np.fmin(
                    hasil_fuzzifikasi['permintaan'][aturan[0]],
                    np.fmin(
                        hasil_fuzzifikasi['persediaan'][aturan[1]],
                        np.fmin(
                            hasil_fuzzifikasi['harga_bahan_baku'][aturan[2]],
                            hasil_fuzzifikasi['tenaga_kerja'][aturan[3]],
                        )
                    )
                )

                # Komposisi Aturan
                if aturan[4] == 'SEDIKIT':
                  implikasi[0] = np.fmax(miu, implikasi[0])

                elif aturan[4] == 'CUKUP':
                  implikasi[1] = np.fmax(miu, implikasi[1])

                else :
                  implikasi[2] = np.fmax(miu, implikasi[2])

                if miu != 0:
                  # Logging: tampilkan detail derajat dan hasil MIN
                  print(f"[R{i}] IF permintaan['{pmt}'] AND persediaan['{psd}'] AND harga_bahan_baku['{bbk}'] AND tenaga_kerja['{tkg}'] THEN produksi['{output}']")
                  print(
                      f"μ{i} = min("
                      f"{hasil_fuzzifikasi['permintaan'][aturan[0]]}, "
                      f"{hasil_fuzzifikasi['persediaan'][aturan[1]]}, "
                      f"{hasil_fuzzifikasi['harga_bahan_baku'][aturan[2]]}, "
                      f"{hasil_fuzzifikasi['tenaga_kerja'][aturan[3]]}, "
                      f") = {miu}"
                  )
                  print("\n")

                MIU.append(miu)
                i +=1
```

## Komposisi Aturan

```python
komposisi_aturan = []
labels_output = list(fungsi_keanggotaan['jumlah_produksi_aktual'].keys())

for i, label in enumerate(labels_output):
    komposisi = np.fmin(
        fungsi_keanggotaan['jumlah_produksi_aktual'][label],
        implikasi[i]
    )
    komposisi_aturan.append(komposisi)

# Agregasi dengan np.fmax.reduce agar dinamis
komposisi_aturan1 = np.fmax.reduce(komposisi_aturan)

print(f"Hasil Komposisi Aturan:")
print(f"μSedikit: {implikasi[0]}")
print(f"μCukup: {implikasi[1]}")
print(f"μBanyak: {implikasi[2]}")
print("\n")
```

## Proses Perhitungan *Fuzzy* Mamdani

```python
def fuzzifikasi(input):
    hasil = {}
    for var, nilai in input.items():
        hasil[var] = {}
        for label, mf in fungsi_keanggotaan[var].items():
            derajat = fuzz.interp_membership(x[var], mf, nilai)
            hasil[var][label] = round(derajat,4)
    return hasil

def aplikasi_fungsi_implikasi(hasil_fuzzifikasi):
    label_skor1 = {'RENDAH': 0, 'SEDANG': 2, 'TINGGI': 3}
    label_skor2 = {'SEDIKIT': 1, 'CUKUP': 0.5, 'BANYAK': 0,
                   'MURAH':1, 'NORMAL':0.5, 'MAHAL':0}
    label_skor_tenagakerja = {'SEDIKIT': 0, 'CUKUP': 1, 'BANYAK': 2}

    label_1 = ['RENDAH', 'SEDANG', 'TINGGI']
    label_2 = ['SEDIKIT', 'CUKUP', 'BANYAK']
    label_3 = ['MURAH', 'NORMAL', 'MAHAL']

    i = 1
    implikasi = [0,0,0]
    MIU = []

    print(f"Aplikasi Fungsi Implikasi Fuzzy:")
    for pmt in label_1:
        for psd in label_2:
            for bbk in label_3:
                for tkg in label_2:
                    # Hitung total skor dari kombinasi label
                    # Konsekuen
                    if pmt == 'RENDAH':
                        total_score = (
                            label_skor2[psd] + label_skor_tenagakerja[tkg] +
                            label_skor2[bbk]
                        )
                        if total_score <= 3:
                          output = 'SEDIKIT'
                        else:
                          output = 'CUKUP'

                    elif pmt == 'SEDANG':
                        total_score = (
                          label_skor1[pmt] + label_skor2[psd] +
                          label_skor2[bbk] + label_skor_tenagakerja[tkg]
                        )
                        if total_score <= 3:
                            output = 'SEDIKIT'
                        elif total_score <= 5:
                            output = 'CUKUP'
                        else:
                            output = 'BANYAK'
                    else:
                        total_score = (
                          label_skor1[pmt] + label_skor2[psd] +
                          label_skor2[bbk] + label_skor_tenagakerja[tkg]
                        )
                        if total_score <= 2:
                            output = 'SEDIKIT'
                        elif total_score <= 4:
                            output = 'CUKUP'
                        else:
                            output = 'BANYAK'

                    # Aturan
                    aturan = [pmt, psd, bbk, tkg, output]
                    hasil = hasil_fuzzifikasi

                    # Aplikasi fungsi implikasi tiap aturan
                    miu = np.fmin(
                        hasil_fuzzifikasi['permintaan'][aturan[0]],
                        np.fmin(
                            hasil_fuzzifikasi['persediaan'][aturan[1]],
                            np.fmin(
                                hasil_fuzzifikasi['harga_bahan_baku'][aturan[2]],
                                hasil_fuzzifikasi['tenaga_kerja'][aturan[3]],
                            )
                        )
                    )

                    # Komposisi Aturan
                    if aturan[4] == 'SEDIKIT':
                      implikasi[0] = np.fmax(miu, implikasi[0])

                    elif aturan[4] == 'CUKUP':
                      implikasi[1] = np.fmax(miu, implikasi[1])

                    else :
                      implikasi[2] = np.fmax(miu, implikasi[2])

                    if miu != 0:
                      # Logging: tampilkan detail derajat dan hasil MIN
                      print(f"[R{i}] IF permintaan['{pmt}'] "
                            f"AND persediaan['{psd}'] "
                            f"AND harga_bahan_baku['{bbk}'] "
                            f"AND tenaga_kerja['{tkg}'] "
                            f"THEN produksi['{output}']")
                      print(
                          f"μ{i} = min("
                          f"{hasil_fuzzifikasi['permintaan'][aturan[0]]}, "
                          f"{hasil_fuzzifikasi['persediaan'][aturan[1]]}, "
                          f"{hasil_fuzzifikasi['harga_bahan_baku'][aturan[2]]}, "
                          f"{hasil_fuzzifikasi['tenaga_kerja'][aturan[3]]}, "
                          f") = {miu}"
                      )
                      print("\n")

                    MIU.append(miu)
                    i +=1
    return MIU, implikasi

def komposisi_aturan_fuzzy(fungsi_keanggotaan, implikasi):
    komposisi_aturan = []
    labels_output = list(fungsi_keanggotaan['jumlah_produksi_aktual'].keys())

    for i, label in enumerate(labels_output):
        komposisi = np.fmin(
            fungsi_keanggotaan['jumlah_produksi_aktual'][label],
            implikasi[i]
        )
        komposisi_aturan.append(komposisi)

    # Agregasi dengan np.fmax.reduce agar dinamis
    komposisi_aturan1 = np.fmax.reduce(komposisi_aturan)

    print(f"Hasil Komposisi Aturan:")
    print(f"μSedikit: {implikasi[0]}")
    print(f"μCukup: {implikasi[1]}")
    print(f"μBanyak: {implikasi[2]}")
    print("\n")

    return komposisi_aturan1

def sistem_inferensi_fuzzy_mamdani(input) :
    # Fuzzifikasi
    hasil_fuzzifikasi = fuzzifikasi(input)

    # Hasil Fuzzifikasi
    for var, memberships in hasil_fuzzifikasi.items():
      print(f"\nFuzzifikasi {var.capitalize()} (nilai = {input[var]}):")
      for label, degree in memberships.items():
          print(f"  {label:<7}: {degree}")
    print("\n")

    # Aplikasi Fungsi Implikasi
    MIU, implikasi = aplikasi_fungsi_implikasi(hasil_fuzzifikasi)

    # Komposisi Aturan
    komposisi_aturan1 = komposisi_aturan_fuzzy(fungsi_keanggotaan,implikasi)

    # Defuzzifikasi
    hasil_defuzzifikasi = fuzz.defuzz(x['jumlah_produksi_aktual'],
                                      komposisi_aturan1,'centroid')

    print(f"Hasil defuzzifikasi: {hasil_defuzzifikasi:.5f}")
    print("\n")

    # Grafik Komposisi Aturan
    plt.plot(x['jumlah_produksi_aktual'],
             fungsi_keanggotaan['jumlah_produksi_aktual']['SEDIKIT'],
             label='SEDIKIT')
    plt.plot(x['jumlah_produksi_aktual'],
             fungsi_keanggotaan['jumlah_produksi_aktual']['CUKUP'],
             label='CUKUP')
    plt.plot(x['jumlah_produksi_aktual'],
             fungsi_keanggotaan['jumlah_produksi_aktual']['BANYAK'],
             label='BANYAK')
    plt.fill_between(x['jumlah_produksi_aktual'],
                     komposisi_aturan1, alpha=0.9, color='grey')

    plt.title('Daerah Hasil Komposisi Aturan Jumlah Produksi (Fuzzy Mamdani)')
    plt.xlabel('Jumlah Produksi')
    plt.ylabel('Derajat Keanggotaan (μ)')
    plt.legend()
    plt.show()

    # Output Akhir
    print(f"Hasil Prediksi Jumlah Produksi:")
    print(f"Permintaan = {input['permintaan']}")
    print(f"Persediaan = {input['persediaan']}")
    print(f"Harga Bahan Baku = {input['harga_bahan_baku']}")
    print(f"Tenaga Kerja = {input['tenaga_kerja']}")
    print(f"Prediksi Jumlah Produksi = {hasil_defuzzifikasi:.5f}")
```
